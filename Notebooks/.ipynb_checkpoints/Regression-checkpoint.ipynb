```python
from sklearn.model_selection import train_test_split
X, y = data2.drop(['pop', 'counties'], axis=1), data2['pop']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
trainingPred = lr.predict(X_train)
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print('r2 socre is on testing data: {}'.format(r2_score(pred, y_test)))
print('r2 socre is on training data: {}'.format(r2_score(trainingPred, y_train)))
```

```python
# Scaled
from sklearn.preprocessing import StandardScaler
X, y = data2.drop(['pop', 'counties'], axis=1), data2['pop']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 3)

scaler = StandardScaler()
X_train= pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns) 
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

lrScaled = LinearRegression()
lrScaled.fit(X_train, y_train)

trainingPred = lrScaled.predict(X_train)
pred = lrScaled.predict(X_test)

print('r2 socre is on testing data: {}'.format(r2_score(pred, y_test)))
print('r2 socre is on training data: {}'.format(r2_score(trainingPred, y_train)))
```

```python
from sklearn.model_selection import cross_val_score
X, y = data2.drop(['pop', 'counties'], axis=1), data2['pop']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 3)
lrCV = LinearRegression()
scores = cross_val_score(lrCV, X_train, y_train, scoring='r2', cv=5)
scores
```

```python
# Feature Selection using RFE
from sklearn.feature_selection import RFE
n_features_optimal = 10

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, n_features_to_select=n_features_optimal)             
rfe = rfe.fit(X_train, y_train)

# predict prices of X_test
y_pred = lm.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
# Feature Selection with RFE and GridSearchCV
from sklearn.model_selection import GridSearchCV
param_grid =  {'n_features_to_select': list(range(1,22))}
folds = 5
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm)
model_cv = GridSearchCV(estimator = rfe, 
                        param_grid = param_grid, 
                        scoring= 'r2', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True)
model_cv.fit(X_train, y_train)
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
plt.figure(figsize=(16,6))

plt.plot(cv_results["param_n_features_to_select"], cv_results["mean_test_score"])
plt.plot(cv_results["param_n_features_to_select"], cv_results["mean_train_score"])
plt.xlabel('number of features')
plt.ylabel('r-squared')
plt.title("Optimal Number of Features")
plt.legend(['test score', 'train score'], loc='upper left')
```

```python
# Best Regression Model
X, y = data2.drop(['pop', 'counties'], axis=1), data2['pop']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 3)

lr = LinearRegression()
lr.fit(X_train, y_train)

trainingPred = lr.predict(X_train)
pred = lr.predict(X_test)

print('r2 socre is on testing data: {}'.format(r2_score(pred, y_test)))
print('r2 socre is on training data: {}'.format(r2_score(trainingPred, y_train)))
```

```python
output = X_test
output['pred'] = pred
output = output.merge(care, how = 'inner', on =  ['none_care', 'first_trimester'
                                                  , 'second_trimester', 'third_trimester',
                                                 'unknown_care', 'year'])
output = output.iloc[: , 1:]
output.to_csv('../Output/regression.csv', index=False)
```
