```python
# imports
import databaseUtils as db
import ioFunctions as fnc
import machineLearningUtils as mlu
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Jupyter
%load_ext autoreload
%autoreload 2
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
def Xy(df, target):
    return df.loc[:, df.columns!=target], df[target]

def testLR(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=31)
    model = LogisticRegression(solver='liblinear').fit(X_train, y_train)
    return model, X_train, X_test, y_train, y_test

def printLR(X, y):
    model, X_train, X_test, y_train, y_test = testLR(X, y)
    mlu.printReport(model, X_train, X_test, y_train, y_test)
    return model

def test(df, target):
    X, y = Xy(df.dropna(), target)
    
    print('===== UNSCALED =====')
    model = printLR(X, y)
    print(model.intercept_, model.coef_)
    
    print('\n===== SCALED =====')
    model = printLR(pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns), y)
    print(model.intercept_, model.coef_)

def testXy(X, y):
    print('===== UNSCALED =====')
    model = printLR(X, y)
    print(model.intercept_, model.coef_)
    
    print('\n===== SCALED =====')
    model = printLR(pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns), y)
    print(model.intercept_, model.coef_)

def extractSelectedColumns(best, X):
    # Extract best k columns
    mask = best.get_support() # for each column, the value is True if k best, False otherwise
    f = [] 
    for is_in_k_best, feature in zip(mask, X.columns):
        if is_in_k_best:
            f.append(feature)

    # Construct a new X
    return pd.DataFrame(X, columns=f)
```

```python
# Univariate Correlation
from scipy.stats.stats import pearsonr
from IPython.display import display

# If pvalue = False, then it will compare by abs(correlation)
def univariateCorrelation(X, y, pvalue=True, threshold=0.01):
    features = list(X)
    correlation = []
    significance = []
    for feature in features:
        correl = pearsonr(X[feature].values, y.values)
        correlation.append(correl[0])
        significance.append(correl[1])
    df = pd.DataFrame()
    df['feature'] = features
    df['correlation'] = correlation
    df['abs_correlation'] = np.abs(correlation)
    df['significance'] = significance
    if pvalue:
        df['significant'] = df['significance'] < threshold
    else:
        df['significant'] = df['abs_correlation'] > threshold
    display(df)

    # Extract best k columns
    mask = df.significant # for each column, the value is True if k best, False otherwise
    f = [] 
    for is_in_k_best, feature in zip(mask, X.columns):
        if is_in_k_best:
            f.append(feature)

    # Construct a new X
    return pd.DataFrame(X, columns=f)
```

```python
X, y = Xy(dataB, target)

# Keep univariate correlated values with p-value > threshold
p = 0.01
X_univ_corr = univariateCorrelation(X, y, p)
testXy(X_univ_corr, y)
```

```python
# Removing Univariate Correlated Values < Threshold
X, y = Xy(dataB, target)

# Keep univariate correlated values >= threshold
cor = 0.1
X_univ_corr = univariateCorrelation(X, y, pvalue=False, threshold=cor)
testXy(X_univ_corr, y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = 5
knn = KNeighborsClassifier(n_neighbors=k)

from sklearn.model_selection import cross_val_score
print('Training accuracy (unscaled): {:0.2%}'.format(cross_val_score(knn, X_train,        y_train, cv=k, scoring=scoring).mean()))
print('Training accuracy (scaled):   {:0.2%}'.format(cross_val_score(knn, X_train_scaled, y_train, cv=k, scoring=scoring).mean()))
```

```python
k_min = 1
k_max = 20
k_range = range(k_min, k_max+1) # [min, max)
k_scores = []
k_scores_scaled = []

# Unscaled
for kn in k_range: # we already have a k for folds, so kn for k neighbors
    knn = KNeighborsClassifier(n_neighbors=kn)
    scores = cross_val_score(knn, X_train, y_train, cv=k, scoring=scoring)
    k_scores.append(scores.mean())

# Scaled
for kn in k_range: # we already have a k for folds, so kn for k neighbors
    knn = KNeighborsClassifier(n_neighbors=kn)
    scores = cross_val_score(knn, X_train_scaled, y_train, cv=k, scoring=scoring)
    k_scores_scaled.append(scores.mean())
    
# Plot
print('===== UNSCALED =====')
plt.plot(k_range, k_scores)
plt.xlabel('(Value of K for KNN)')
plt.ylabel('Cross-Validated Accuracy')

# Find the highest accuracy furthest to the right (larger k ~ reduced complexity)
scores = pd.DataFrame(k_scores, index=pd.RangeIndex(start=k_min, stop=k_max+1, step=1), columns={'Score'}).sort_values(by='Score', ascending=False)
print(scores.head(10).transpose())

print('===== SCALED =====')
plt.plot(k_range, k_scores_scaled)
plt.xlabel('(Value of K for KNN)')
plt.ylabel('Cross-Validated Accuracy')
plt.legend(['Unscaled', 'Scaled'])

# Find the highest accuracy furthest to the right (larger k ~ reduced complexity)
scores_scaled = pd.DataFrame(k_scores_scaled, index=pd.RangeIndex(start=k_min, stop=k_max+1, step=1), columns={'Score'}).sort_values(by='Score', ascending=False)
print(scores_scaled.head(10).transpose())
```
